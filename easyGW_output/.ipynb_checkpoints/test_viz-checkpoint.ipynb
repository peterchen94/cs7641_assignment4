```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
fileList = os.listdir('.')
```

```python
plt.figure()
legend_list=[]
for file in fileList:
    if ('Value' in file or 'Policy' in file) and file.endswith('csv'):
        legend_list.append(file)
        df=pd.read_csv(file)
        plt.plot(df['reward'])
plt.legend(legend_list)
```

```python
legend_list=[]
for file in fileList:
    if ('Q' in file) and file.endswith('csv'):
        plt.figure()
        legend_list.append(file)
        df=pd.read_csv(file)
        plt.plot(df['reward'])
        plt.title(file)
        plt.xlim(left=0,right = 100)
        plt.ylim((-400,100))
```
